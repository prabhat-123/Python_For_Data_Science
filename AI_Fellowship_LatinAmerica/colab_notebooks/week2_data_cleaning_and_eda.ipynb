## Pandas Data Types

Pandas provides several data types to handle different types of data efficiently. These data types are associated with the columns (or Series) of a dataframe. Understanding and specifying the correct data types is crucial for data manipulation and analysis.

### Numeric Data Types

1. **int64:** This data type represents signed 64-bit integers. It is suitable for handling whole numbers.

2. **float64:** This data type represents 64-bit floating-point numbers. It is used for decimal numbers and numbers with fractional parts.

### Categorical Data Types

3. **category:** Categorical data type is used to represent categorical variables with a fixed, predefined set of values. It can significantly reduce memory usage and improve performance when dealing with categorical data.

### Datetime Data Types

4. **datetime64:** This data type is used for date and time values. It allows for easy manipulation of dates and times.

### Text Data Types

5. **object:** The object data type is a generic data type used for columns that contain mixed data types or text. While it's flexible, it may not be as memory-efficient as other specific data types.

### Boolean Data Type

6. **bool:** This data type represents Boolean values, i.e., True or False. It's used for binary data or conditions.

# Data Cleaning

Data cleaning, also known as data cleansing or data scrubbing, is an essential step in the data preprocessing pipeline. It involves identifying and correcting errors, inconsistencies, and inaccuracies in a dataset to ensure that the data is accurate, reliable, and suitable for analysis. Proper data cleaning is critical because the quality of your analysis and insights depends on the quality of the data you work with.

## Key Processes in Data Cleaning:

1. **Removing Duplicates:**
   - **Identification:** Detect and identify duplicate rows or records in the dataset.
   - **De-duplication:** Decide whether to remove duplicates to prevent skewing of analysis results.


2. **Handling Missing Data:**
   - **Identification:** Identify missing values (e.g., NaN or NULL) in the dataset.
   - **Imputation:** Decide how to handle missing data, which may involve imputing missing values.
   - **Deletion:** In some cases, it may be necessary to remove rows or columns with a significant amount of missing data.

3. **Standardizing Data:**
   - **Data Types:** Ensure that data types are consistent and appropriate for each column.
   - **Encoding:** Convert categorical variables into numerical form through encoding techniques.

4. **Handling Outliers:**
   - **Identification:** Identify outliers through statistical methods or visualizations.
   - **Treatment:** Decide whether to remove, transform, or keep outliers based on domain knowledge.

5. **Correcting Inaccuracies:**
   - **Data Validation:** Validate data against predefined rules or constraints.
   - **Data Transformation:** Correct inaccuracies, inconsistencies, or erroneous entries based on domain knowledge.

6. **Handling Data Integrity Issues:**
   - **Referential Integrity:** Ensure that relationships between tables (if applicable) are maintained.
   - **Data Integration:** If the data comes from multiple sources, integrate it while resolving any conflicts or inconsistencies.


## Importance of Data Cleaning:

- **Data Quality:** Ensures high-quality and reliable data for accurate analysis.
- **Analysis Accuracy:** Leads to more accurate and trustworthy analysis results.
- **Reducing Bias:** Reduces bias and errors introduced by erroneous or missing data.
- **Data Governance:** Contributes to data governance efforts by ensuring data integrity and compliance.

<li>In summary, data cleaning is a critical step in the data preparation process, significantly impacting the quality and reliability of data analysis and insights.</li>

<li>Properly cleaned data is the foundation for making informed decisions and deriving meaningful conclusions from datasets.</li>

## Missing Values in Data

<li><b>Missing values</b>, also known as <b>missing data</b>, refer to the absence of data or information for one or more variables or observations in a dataset.</li>

<li>They are a common occurrence in real-world datasets and can have a significant impact on data analysis and modeling.</li>

<li>Understanding the sources of missing data is crucial for handling them effectively.</li>

### Sources of Missing Data

- **Non-Response:** Missing data due to respondents not providing information, often driven by privacy concerns, time constraints, or optional questions.

- **Data Entry Errors:** Missing data resulting from mistakes during data entry, including typographical errors, scanning or recording issues, and misinterpretation.

- **Measurement Limitations:** Missing data due to limitations in measurement processes, such as instrument failures, incompatibility, or human errors during data collection.

### Handling Missing Values

Handling missing values is essential to avoid biased analyses and maintain data integrity. Strategies for dealing with missing data include:

- **Imputation:** Replacing missing values with estimates or predictions. Common imputation methods include mean imputation, median imputation, and regression imputation.

- **Deletion:** Removing rows or columns with missing values. This approach is suitable when missing values are random and do not significantly affect the dataset's integrity.

- **Interpolation:** Estimating missing values based on adjacent or similar observations. Time series data, for instance, can be interpolated to fill gaps.

- **Advanced Techniques:** Machine learning models, such as K-nearest neighbors (KNN) or matrix factorization methods, can be used for imputing missing values.

The choice of strategy depends on the nature of the data, the extent of missing values, and the goals of the analysis. Handling missing data appropriately ensures that the resulting analyses and models are accurate and reliable.

## Relation Between Stroke & Hypetension

```python
stroke_hypertension = pd.crosstab(index=df["hypertension"], columns = df["stroke"])
stroke_hypertension.columns = ["not_stroke", "stroke"]
stroke_hypertension.reset_index(inplace=True)
```

```python
plt.figure(figsize=(8, 6))
ax = stroke_hypertension.plot(x='hypertension', kind='bar', stacked=True)
plt.show()
```
